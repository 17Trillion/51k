```python
globaleconomy = 137 #trillion    #1400 would be 1.4 Quadrillion
inflationrate = 7.5 #percent


def ubicalc(ubi):
    upotril = ubi * 330000000/1000000000000  #ubi * population per trillion to ratio against the globaleconomy trillions as 1s.
    multiplierlist =[(globaleconomy+(upotril*x))/(globaleconomy+(upotril*(x+1))) for x in range(101)] 
    #(startingGlobaleconomy+prioryears'UBI)/(startingGlobaleconomy+thisyear'sUBI)
    
    acum = 1
    flist = []
    ubivalist =[]
    for mult in multiplierlist:
        acum = acum * mult
        flist.append(acum)  #list of percents multiplied over 100 years, shows percentage of starting value over time.
        ubivalist.append(acum*ubi)  #all years multiplied, value of initial ubi through each year stored in ubivalist
    return flist,ubivalist

ubiinput=51000 #number chosen to optimize for benefit vs initial effects

flist,ubivalist = ubicalc(ubiinput)

n=0

print("percent" +" || "+ "yr" +" || "+ "UBI min value" +"\n")
for i in range(len(flist)):
    
    print('{:.5f}'.format(flist[i]) +" || "+ '{:>2}'.format(n) +" || "+'{:.2f}'.format(ubivalist[i]) +"\n")
    n=n+1
```

```python
inflation_rate = 1-(inflationrate/100)  # .925 = 7.5% inflation .962=3.8% (100 yr avg)
current_inflation51=[]
for i in range(101):
    current_inflation51.append(51000*(inflation_rate**i))  #effects of inflation on 51k dollars over time
```

```python
ubilist=[3000,8000,12000,24000,51000,64000,80000,120000,240000]  #fill with different UBIs you want to see graphed
fl_list=[]
uv_list=[]

for u in ubilist:
    fll,ull = ubicalc(u)
    fl_list.append(fll)
    uv_list.append(ull)
```

```python
combolist = []
for i in range(len(uv_list[4])):
    combolist.append(uv_list[4][i]*(inflation_rate**i))
```

```python
import matplotlib.pyplot as plt
import numpy as np

x= range(0,101)

fig, ax = plt.subplots(figsize=(15,12))

for y in uv_list:
    ax.plot(x, y, linewidth=1.0, color ='orange')   #range of other potential UBI options
ax.plot(x, uv_list[4], linewidth=1.5, color ='blue')  #51k is blue if [4] in uv_list
ax.plot(x, uv_list[8], linewidth=1.0, color ='cyan')  
ax.plot(x, current_inflation51, linewidth=1,color='red') #current inflation working on a starting value of 51k
#ax.plot(x, combolist, linewidth=1,color='purple')
#ax.set_yscale('log')   #<-- turn this on to see this graph in logarithmic scale.

plt.show()
```
